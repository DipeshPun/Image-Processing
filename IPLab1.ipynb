```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
#read images
img1= cv2.imread("/content/drive/MyDrive/images-20250302T053354Z-001/images/everest.jpg",0)
img1=cv2.resize(img1,(300,300))
print(img1)
print(img1.shape)
```

```python
plt.imshow(img1,cmap='gray')
plt.title("Everest")
plt.show()
```

```python
img2= cv2.imread("/content/drive/MyDrive/images-20250302T053354Z-001/images/flag.jpg",0)
img2=cv2.resize(img2,(300,300))
print(img2)
print(img2.shape)
```

```python
plt.imshow(img2,cmap='gray')
plt.title("Flag")
plt.show()
```

```python
#Adding two images together
img3 = cv2.addWeighted(img1, 0.3, img2, 0.7, 0)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.title("Addition")
plt.imshow(img3, cmap='gray')
plt.show()
```

```python
#Adding two images to form new image
import numpy as np

img1 = np.array([[100, 200, 123],
              [45, 145, 166]])
img2 = np.array([[136, 133, 155],
              [126, 147, 188]])

img3 = np.empty(img1.shape)

for i in range(img1.shape[0]):
  for j in range(img1.shape[1]):
    img3[i, j] = img1[i, j]*0.3 + img2[i, j]*0.7

print(img3)
```

```python

imgone =  cv2.imread("/content/drive/MyDrive/images-20250302T053354Z-001/images/everest.jpg")
imgone = cv2.resize(imgone, (250, 250))
imgthree = np.empty((250, 250, 3), dtype=np.uint8)
for i in range(0, 250):
    for j in range(0, 250):
        imgthree[i][j] = np.clip(imgone[i][j] *1.5, 0, 255)
plt.imshow(cv2.cvtColor(imgthree, cv2.COLOR_BGR2RGB))
```

```python
output_path='/content/drive/MyDrive/images-20250302T053354Z-001/images/updagedeverst.jpg'
cv2.imwrite(output_path, imgthree)
```

```python
show_image=cv2.imread(output_path)
plt.imshow(cv2.cvtColor(show_image,cv2.COLOR_BGR2RGB))
```
